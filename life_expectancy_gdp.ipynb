```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("all_data.csv")
df.head()
```

```python
df.info()
```

```python
print(df.Country.unique())
print(df.Year.unique())
```

```python
df = df.rename(columns={'Country': 'country', 'Year': 'year', 'Life expectancy at birth (years)': 'life_expectancy', 'GDP':'gdp'})
df.head()
```

```python
df_by_country = df.groupby('country').mean().reset_index()
df_by_country = df_by_country.drop('year', axis=1)
df_by_country.head()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='country', y='life_expectancy', data=df)
plt.title('Life Expectancy at Birth by Country')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='country', y='gdp', data=df)
plt.title('GDP in Trillions by Country')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.violinplot(x='country', y='life_expectancy', data=df)
plt.title('Life Expectancy Distribution by Country')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=df, y='life_expectancy', x='gdp', hue='country')
plt.title('GDP vs. Life Expectancy by Country')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x=df.year, y=df.gdp, hue=df.country)
plt.title('GDP By Year')
plt.legend(df_by_country.country, loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x=df.year, y=df.life_expectancy, hue=df.country)
plt.title('Life Expectancy By Year')
plt.legend(df_by_country.country, loc='upper left')
plt.show()
```
